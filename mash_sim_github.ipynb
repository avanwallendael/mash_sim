```python
print(toytree.__version__)
```

```python
treebig = toytree.rtree.unittree(50, treeheight=1e6, seed=123)
```

```python
# draw tree showing idx labels
treebig.draw(tree_style='p');
```

```python
import ipcoal
```

```python
# initialize a model object given a species tree and Ne setting.
model = ipcoal.Model(tree=treebig, Ne=1e5)
```

```python
# create a Model and simulate a 100bp locus
modela = ipcoal.Model(treebig, Ne=1e5, recomb=1e-9, seed=123)
modela.sim_loci(nloci=1000, nsites=100)
```

```python
# the results table
modela.df
```

```python
# concatenate and write only some loci to test
modela.write_concat_to_phylip(
    outdir="~/Desktop/ipcoal-sims/sims", name="locs-1-3.phy", idxs=[1,2,3]
)
```

```python
# write all loci 
modela.write_loci_to_phylip(outdir="~/Desktop/ipcoal-sims/sims", name_prefix="simbig_a")
```
